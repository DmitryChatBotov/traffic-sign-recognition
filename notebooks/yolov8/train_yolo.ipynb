```python
import gdown
import tarfile 
import torch
import random
import os
import torch
import numpy as np
from ultralytics import YOLO
import optuna
```

```python
RANDOM_SEED=42
random.seed(RANDOM_SEED)
os.environ["PYTHONHASHSEED"] = str(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED)
```

```python
url = "https://drive.google.com/uc?id=1Gej_GgjyBe79lyZqwtIwg3mXxPN-SloC"
path = 'traffic-sign'
gdown.download(url, path)
```

```python
with tarfile.open('traffic-sign') as dataset:
    dataset.extractall('.')
```

```python
model = YOLO('yolov8s.pt')
result = model.train(data='traffic-signs.yaml', epochs=15, batch=24, seed=RANDOM_SEED)
```
