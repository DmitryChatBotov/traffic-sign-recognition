```python
import torch
import random
import os
import torch
import numpy as np
from ultralytics import YOLO
```

–§–∏–∫—Å–∏—Ä—É–µ–º random_seed –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏

```python
RANDOM_SEED=42
random.seed(RANDOM_SEED)
os.environ["PYTHONHASHSEED"] = str(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED)
```

–û–±—É—á–∞–µ–º –Ω–µ–π—Ä–æ—Å–µ—Ç—å

```python
torch.cuda.empty_cache()
```

```python
model = YOLO('yolov8s.pt')
```

```python
epochs = 15
batch=24
image_size = 640
```

```python
result = model.train(data='./output_data/traffic-sign.yaml', epochs=epochs, batch=batch, seed=RANDOM_SEED, imgsz=image_size)
```

```python
metrics = model.val()
```
